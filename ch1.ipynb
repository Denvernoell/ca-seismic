```python
import sympy.physics.units as u

u.find_unit("pound")
u.util.quantity_simplify(1000 * u.pound / (u.inch**2))
from sympy import symbols

# K = 3*E*I / (H**3)
E, I, H = symbols("E I H")
K = 3 * E * I / (H**3)
K
```

```python
K.subs(
    {
        E: 29_000 * (1000 * u.pound / u.inch**2),
        I: 12_500 * u.inch**4,
        H: 12 * u.inch,
    }
)
```
